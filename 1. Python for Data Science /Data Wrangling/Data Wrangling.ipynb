```python
# ============================================================
# 1. IMPORT LIBRARIES
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

!pip install pyfiglet
from pyfiglet import figlet_format

print("Libraries Imported Successfully!")

# ============================================================
# 2. CREATE / IMPORT SAMPLE CSV DATA
# ============================================================
# (You can replace this with pd.read_csv('file.csv') if needed)

data = {
    "Name": ["Amit", "Sita", "Rohit", "Priya", "Kiran", "Neha", "Vikas", "Sonam"],
    "Age": [23, 29, 21, 32, 27, 24, 35, 30],
    "City": ["Delhi", "Mumbai", "Delhi", "Bangalore", "Mumbai", "Delhi", "Bangalore", "Mumbai"],
    "Salary": [35000, 52000, 30000, 65000, 48000, 40000, 75000, 60000]
}

df = pd.DataFrame(data)
df.to_csv("employee-data.csv", index=False)

print("\nDataset Created Successfully!\n")
print(df)

# ============================================================
# 3. IMPORT CSV USING PANDAS
# ============================================================
df = pd.read_csv("employee-data.csv")
print("\nCSV Loaded Successfully!")
print(df.head())

# ============================================================
# 4. HANDLING COLUMNS (ADD, RENAME, DROP)
# ============================================================
df["Annual_Salary"] = df["Salary"] * 12
df.rename(columns={"Name": "Employee_Name"}, inplace=True)

# Drop Example (Not dropping now but demo)
# df.drop("Age", axis=1, inplace=True)

print("\nUpdated Columns:")
print(df.head())

# ============================================================
# 5. FILTERING ROWS
# ============================================================
high_salary = df[df["Salary"] > 50000]
print("\nEmployees With Salary > 50,000:")
print(high_salary)

# ============================================================
# 6. SORTING DATA
# ============================================================
sorted_df = df.sort_values(by="Salary", ascending=False)
print("\nSorted Data by Salary (Descending):")
print(sorted_df)

# ============================================================
# 7. GROUPING DATA
# ============================================================
grouped = df.groupby("City")["Salary"].mean().reset_index()
print("\nAverage Salary by City:")
print(grouped)

# ============================================================
# 8. MERGING TWO DATAFRAMES
# ============================================================
extra_data = {
    "City": ["Delhi", "Mumbai", "Bangalore"],
    "Region_Code": ["DL", "MB", "BL"]
}

df_region = pd.DataFrame(extra_data)
merged_df = pd.merge(df, df_region, on="City", how="left")

print("\nMerged Data:")
print(merged_df)

# ============================================================
# 9. DATA VISUALIZATION
# ============================================================

# BAR CHART: Average Salary by City
plt.figure(figsize=(7,5))
plt.bar(grouped["City"], grouped["Salary"])
plt.xlabel("City")
plt.ylabel("Average Salary")
plt.title("Average Salary by City")
plt.grid(True)
plt.show()

# SCATTER PLOT: Age vs Salary
plt.figure(figsize=(7,5))
plt.scatter(df["Age"], df["Salary"])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.grid(True)
plt.show()

# HEATMAP: Correlation
plt.figure(figsize=(6,4))
sns.heatmap(df.select_dtypes(include=np.number).corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# ============================================================
# 10. CUSTOM USER INPUT
# ============================================================
print("\n------ ENTER EMPLOYEE DETAILS ------")
emp_name = input("Enter Employee Name: ")
emp_age = int(input("Enter Age: "))
emp_city = input("Enter City: ")
emp_salary = int(input("Enter Salary: "))

# Create input DataFrame
user_df = pd.DataFrame({
    "Employee_Name": [emp_name],
    "Age": [emp_age],
    "City": [emp_city],
    "Salary": [emp_salary],
    "Annual_Salary": [emp_salary * 12]
})

print("\n------------ FINAL USER DATA -------------")
print(user_df)

# ============================================================
# 11. FIGLET FORMAT OUTPUT
# ============================================================
fig = figlet_format(f"{emp_name}", font="standard")
fig_salary = figlet_format(f" {emp_salary}", font="standard")

print("\n==================== ASCII ART OUTPUT ====================")
print(fig)
print("------------ Salary ------------")
print(fig_salary)
print("==========================================================\n")
```
