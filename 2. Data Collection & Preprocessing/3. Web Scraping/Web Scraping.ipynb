**Step 2: Import Libraries**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
folder_path = "/content/drive/MyDrive/Colab Notebooks/DSE(extra)/2. Data Collection"
os.makedirs(folder_path, exist_ok=True)
```

**Step 3: Send Request to Website**

```python
url = "https://books.toscrape.com/"  # paste link of web scraping data

response = requests.get(url)

# Check if request was successful
print("Status Code:", response.status_code)
```

**Step 4: Parse HTML using BeautifulSoup**

```python
soup = BeautifulSoup(response.text, "html.parser")
```

**Step 5: Extract Product Data**

```python
books = soup.find_all("article", class_="product_pod")

book_name = []
price = []
rating = []
availability = []
```

**Step 6: Loop Through Products**

```python
for book in books:
    # Book Name
    name = book.h3.a["title"]
    book_name.append(name)

    # Price
    book_price = book.find("p", class_="price_color").text
    price.append(book_price)

    # Rating
    book_rating = book.find("p", class_="star-rating")["class"][1]
    rating.append(book_rating)

    # Availability
    avail = book.find("p", class_="instock availability").text.strip()
    availability.append(avail)
```

**Step 7: Store Data in DataFrame**

```python
data = {
    "Book Name": book_name,
    "Price": price,
    "Rating": rating,
    "Availability": availability
}

df = pd.DataFrame(data)
df
```

**Save CSV File to That Path**

```python
file_path = folder_path + "/scraped_products1.csv"

df.to_csv(file_path, index=False)

print("File saved successfully at:")
print(file_path)
```
