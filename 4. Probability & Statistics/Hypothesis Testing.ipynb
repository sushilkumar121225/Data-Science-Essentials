# **Hypothesis Testing**

```python
# HYPOTHESIS TESTING IN PYTHON
# t-test | Chi-Square Test | ANOVA

print("""
H  H  Y   Y  PPPP   OOO   TTTTT  H  H  EEEEE  SSSS  I  SSSS
H  H   Y Y   P   P O   O    T    H  H  E      S     I  S
HHHH    Y    PPPP  O   O    T    HHHH  EEEE   SSSS  I  SSSS
H  H    Y    P     O   O    T    H  H  E         S  I     S
H  H    Y    P      OOO     T    H  H  EEEEE  SSSS  I  SSSS
""")

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load Dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/hypothesis_testing_dataset.csv")
print("\nDataset Preview:")
print(df.head())

# Basic Info
print("\nDataset Information:")
print(df.info())

# 1. T-TEST (Independent Sample)
male_scores = df[df["Gender"] == "Male"]["Exam_Score"]
female_scores = df[df["Gender"] == "Female"]["Exam_Score"]

t_stat, p_val = stats.ttest_ind(male_scores, female_scores)

print("\nT-Test Result:")
print("T-Statistic:", t_stat)
print("P-Value:", p_val)

# Visualization
plt.figure()
plt.hist(male_scores, alpha=0.7, label="Male")
plt.hist(female_scores, alpha=0.7, label="Female")
plt.title("Exam Score Distribution by Gender")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# 2. CHI-SQUARE TEST
contingency_table = pd.crosstab(df["Gender"], df["Department"])
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print("\nChi-Square Test Result:")
print("Chi-Square Value:", chi2)
print("P-Value:", p)

# Visualization
contingency_table.plot(kind="bar")
plt.title("Gender vs Department")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# 3. ANOVA TEST
cs = df[df["Department"] == "CS"]["Exam_Score"]
it = df[df["Department"] == "IT"]["Exam_Score"]
ece = df[df["Department"] == "ECE"]["Exam_Score"]

f_stat, p_anova = stats.f_oneway(cs, it, ece)

print("\nANOVA Test Result:")
print("F-Statistic:", f_stat)
print("P-Value:", p_anova)

# Visualization
plt.figure()
plt.boxplot([cs, it, ece], labels=["CS", "IT", "ECE"])
plt.title("Exam Score Comparison Across Departments")
plt.xlabel("Department")
plt.ylabel("Exam Score")
plt.show()

print("\nAnalysis Completed Successfully ")
```
